# Perceptrones: Fundamentos de las Redes Neuronales

## 1. Introducción a los Perceptrones

El perceptrón es la unidad básica y más antigua de las redes neuronales artificiales. Desarrollado por Frank Rosenblatt en 1957, representa un modelo matemático inspirado en el funcionamiento de las neuronas biológicas.

Un perceptrón toma múltiples entradas (inputs), las procesa y produce una única salida binaria basada en una función de activación.

## 2. Estructura y Componentes

![Estructura de un Perceptrón](https://miro.medium.com/max/1400/1*n6sJ4yZQzwKL9wnF5wnVNg.png)

### Componentes principales:
- **Entradas (x₁, x₂, ..., xₙ)**: Valores numéricos que representan características o atributos.
- **Pesos (w₁, w₂, ..., wₙ)**: Valores que determinan la importancia de cada entrada.
- **Sesgo (bias)**: Un término adicional (b) que permite ajustar el umbral de activación.
- **Función de activación**: Determina la salida del perceptrón según la suma ponderada de entradas.

## 3. Modelo Matemático

La salida de un perceptrón se calcula mediante:

$$y = f(\sum_{i=1}^{n} w_i x_i + b)$$

Donde:
- $x_i$ son las entradas
- $w_i$ son los pesos
- $b$ es el sesgo
- $f$ es la función de activación

### Función de Activación Escalón:

$$f(x) = \begin{cases} 
1 & \text{si } \sum_{i=1}^{n} w_i x_i + b > 0 \\
0 & \text{en caso contrario} 
\end{cases}$$

## 4. Proceso de Aprendizaje

El aprendizaje en un perceptrón consiste en ajustar los pesos y el sesgo para que clasifique correctamente los ejemplos de entrenamiento.

### Algoritmo de Aprendizaje:

1. Inicializar los pesos y el sesgo (generalmente con valores aleatorios pequeños).
2. Para cada ejemplo de entrenamiento:
   - Calcular la salida del perceptrón con los pesos actuales.
   - Comparar la salida con el valor objetivo.
   - Actualizar pesos según la regla: $w_i = w_i + \Delta w_i$, donde $\Delta w_i = \eta (t - y) x_i$
     - $\eta$ es la tasa de aprendizaje (learning rate)
     - $t$ es el valor objetivo
     - $y$ es la salida actual del perceptrón
3. Repetir hasta que el error sea aceptable o se alcance un número máximo de iteraciones.

## 5. Implementación Básica en Python

```python
import numpy as np

class Perceptron:
    def __init__(self, learning_rate=0.01, n_iterations=1000):
        self.learning_rate = learning_rate
        self.n_iterations = n_iterations
        self.weights = None
        self.bias = None
        
    def activation_function(self, x):
        return 1 if x > 0 else 0
    
    def fit(self, X, y):
        # Inicializar pesos y bias
        n_samples, n_features = X.shape
        self.weights = np.zeros(n_features)
        self.bias = 0
        
        # Entrenar el perceptrón
        for _ in range(self.n_iterations):
            for idx, x_i in enumerate(X):
                # Calcular la salida
                linear_output = np.dot(x_i, self.weights) + self.bias
                y_predicted = self.activation_function(linear_output)
                
                # Actualizar pesos y bias
                update = self.learning_rate * (y[idx] - y_predicted)
                self.weights += update * x_i
                self.bias += update
    
    def predict(self, X):
        linear_output = np.dot(X, self.weights) + self.bias
        y_predicted = [self.activation_function(x) for x in linear_output]
        return np.array(y_predicted)
```

## 6. Limitaciones de los Perceptrones

- **Problemas linealmente separables**: Los perceptrones solo pueden resolver problemas donde las clases son separables por una línea recta (o hiperplano en dimensiones superiores).
- **XOR Problem**: No pueden resolver el problema XOR, como demostró Minsky y Papert en 1969.

![XOR Problem](https://miro.medium.com/max/941/1*Tc8UgR_fjI_uHsCGyvr3CQ.png)

## 7. Evolución hacia Redes Neuronales Multicapa

Para superar las limitaciones de los perceptrones simples, se desarrollaron redes con múltiples capas de perceptrones (Perceptrones Multicapa o MLP):

- **Capa de entrada**: Recibe los datos iniciales.
- **Capas ocultas**: Procesan la información mediante funciones de activación no lineales.
- **Capa de salida**: Produce el resultado final.

Las redes multicapa, junto con el algoritmo de retropropagación (backpropagation), permitieron resolver problemas no lineales y dieron origen al campo moderno del Deep Learning.

## 10. Perceptrón Multicapa (MLP)

El Perceptrón Multicapa extiende el concepto del perceptrón simple agregando capas intermedias (ocultas) que permiten resolver problemas no linealmente separables.

### Estructura de un MLP

![Estructura de un MLP](https://miro.medium.com/max/1400/1*3fA77_mLNiJTSgZFhYnU0Q.png)

Un MLP típico contiene:

- **Capa de entrada**: Neuronas que reciben los datos originales.
- **Capas ocultas**: Una o más capas donde ocurre el procesamiento principal.
- **Capa de salida**: Proporciona el resultado final (predicción).

### Funciones de Activación Avanzadas

A diferencia del perceptrón básico que usa la función escalón, los MLPs utilizan funciones de activación diferenciables:

1. **Sigmoide**: $f(x) = \frac{1}{1 + e^{-x}}$
   - Rango entre 0 y 1
   - Útil para probabilidades

2. **Tangente Hiperbólica (tanh)**: $f(x) = \frac{e^x - e^{-x}}{e^x + e^{-x}}$
   - Rango entre -1 y 1
   - Generalmente mejor rendimiento que sigmoide

3. **ReLU (Rectified Linear Unit)**: $f(x) = \max(0, x)$
   - Computacionalmente eficiente
   - Ayuda a mitigar el problema del gradiente desvaneciente
   - Dominante en redes modernas

4. **Leaky ReLU**: $f(x) = \max(\alpha x, x)$ donde $\alpha$ es un valor pequeño (ej. 0.01)
   - Soluciona el "problema de las neuronas muertas" de ReLU

## 11. Algoritmo de Retropropagación (Backpropagation)

El algoritmo de retropropagación es fundamental para entrenar MLPs:

1. **Propagación hacia adelante**: Los datos pasan desde la entrada hasta la salida.
2. **Cálculo del error**: Se compara la salida con el valor esperado.
3. **Propagación hacia atrás**: El error se propaga desde la salida hacia la entrada.
4. **Actualización de pesos**: Se ajustan los pesos para minimizar el error.

Matemáticamente, se basa en la regla de la cadena del cálculo para computar gradientes:

$$\frac{\partial E}{\partial w_{ij}} = \frac{\partial E}{\partial o_j} \cdot \frac{\partial o_j}{\partial net_j} \cdot \frac{\partial net_j}{\partial w_{ij}}$$

Donde:
- $E$ es el error
- $o_j$ es la salida de la neurona $j$
- $net_j$ es la entrada neta de la neurona $j$
- $w_{ij}$ es el peso entre la neurona $i$ y $j$

## 12. Implementación de un MLP en Python con NumPy

```python
import numpy as np

class MLP:
    def __init__(self, layer_sizes, learning_rate=0.1, max_epochs=1000):
        self.layer_sizes = layer_sizes
        self.learning_rate = learning_rate
        self.max_epochs = max_epochs
        self.weights = []
        self.biases = []
        
        # Inicializar pesos y sesgos
        for i in range(len(layer_sizes) - 1):
            self.weights.append(np.random.randn(layer_sizes[i], layer_sizes[i+1]) * 0.1)
            self.biases.append(np.random.randn(1, layer_sizes[i+1]) * 0.1)
    
    def sigmoid(self, x):
        return 1 / (1 + np.exp(-x))
    
    def sigmoid_derivative(self, x):
        return x * (1 - x)
    
    def forward(self, X):
        activations = [X]
        for i in range(len(self.weights)):
            net = np.dot(activations[-1], self.weights[i]) + self.biases[i]
            act = self.sigmoid(net)
            activations.append(act)
        return activations
    
    def backward(self, X, y, activations):
        # Calcular el error
        error = y - activations[-1]
        deltas = [error * self.sigmoid_derivative(activations[-1])]
        
        # Propagar el error hacia atrás
        for i in reversed(range(len(self.weights) - 1)):
            delta = np.dot(deltas[0], self.weights[i+1].T) * self.sigmoid_derivative(activations[i+1])
            deltas.insert(0, delta)
        
        # Actualizar pesos y sesgos
        for i in range(len(self.weights)):
            self.weights[i] += self.learning_rate * np.dot(activations[i].T, deltas[i])
            self.biases[i] += self.learning_rate * np.sum(deltas[i], axis=0, keepdims=True)
    
    def train(self, X, y):
        for epoch in range(self.max_epochs):
            # Forward propagation
            activations = self.forward(X)
            
            # Backward propagation
            self.backward(X, y, activations)
            
            # Calcular error
            if epoch % 100 == 0:
                error = np.mean(np.abs(y - activations[-1]))
                print(f"Epoch {epoch}, Error: {error}")
    
    def predict(self, X):
        return self.forward(X)[-1]
```

## 17. Desafíos en el Entrenamiento de Redes Neuronales

### Problemas del Gradiente

En redes profundas surgen dos problemas principales:

- **Gradiente Desvaneciente (Vanishing Gradient)**: Cuando las derivadas son muy pequeñas, los pesos apenas se actualizan, especialmente en capas iniciales.
- **Gradiente Explosivo (Exploding Gradient)**: Las derivadas son tan grandes que causan actualizaciones desmedidas en los pesos.

**Soluciones**:
- Funciones de activación como ReLU
- Inicialización de pesos (Xavier/Glorot, He)
- Batch Normalization
- Conexiones residuales (ResNet)

### Sobreajuste (Overfitting)

El modelo aprende demasiado bien los datos de entrenamiento y no generaliza a nuevos datos.

**Estrategias para combatirlo**:
- Regularización (L1, L2)
- Dropout
- Early Stopping
- Aumentación de datos
- Reducción de complejidad del modelo

## 18. Hiperparámetros y su Optimización

- **Función de activación**: Afecta la capacidad de aprendizaje

import torch.optim as optim

class MLP(nn.Module):
    def __init__(self, input_dim, hidden_dims, output_dim):

# perceptron con contexto empresarial - crediticio

El modelo simula un sistema automatizado de aprobación de préstamos que una institución financiera podría utilizar para tomar decisiones preliminares sobre solicitudes de crédito basándose en tres factores principales:



````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

class Perceptron:
    def __init__(self, learning_rate=0.01, n_iterations=1000):
        self.learning_rate = learning_rate
        self.n_iterations = n_iterations
        self.weights = None
        self.bias = None
        
    def weighted_sum(self, X):
        """Calcula u = x1*w1 + x2*w2 + x3*w3 + b para cada muestra"""
        return np.dot(X, self.weights) + self.bias
    
    def activate(self, u):
        """Función de activación escalón"""
        return np.where(u > 0, 1, 0)
    
    def fit(self, X, y):
        n_samples, n_features = X.shape
        
        # Inicializar pesos y bias
        self.weights = np.zeros(n_features)
        self.bias = 0
        
        # Historial de precisión para visualización
        self.accuracy_history = []
        
        # Entrenamiento del perceptrón
        for i in range(self.n_iterations):
            # Calcula la suma ponderada
            u = self.weighted_sum(X)
            
            # Aplica la función de activación
            y_pred = self.activate(u)
            
            # Actualiza pesos y bias cuando hay error
            errors = y - y_pred
            self.weights += self.learning_rate * np.dot(X.T, errors)
            self.bias += self.learning_rate * np.sum(errors)
            
            # Calcula precisión y la guarda
            accuracy = np.mean(y == y_pred)
            self.accuracy_history.append(accuracy)
            
            # Si alcanza 100% de precisión, termina antes
            if accuracy == 1.0:
                break
                
    def predict(self, X):
        u = self.weighted_sum(X)
        return self.activate(u)

# Generamos datos simulados para un caso de aprobación de préstamos
np.random.seed(42)
n_samples = 100

# Características: 
# x1: Ingresos (en miles)
# x2: Historial crediticio (puntuación)
# x3: Años de empleo

# Generar datos
ingresos = np.random.normal(50, 15, n_samples)
historial_crediticio = np.random.normal(650, 100, n_samples)
anios_empleo = np.random.normal(5, 3, n_samples)

# Combinar características
X = np.column_stack((ingresos, historial_crediticio, anios_empleo))

# Generar etiquetas (aprobado=1, rechazado=0) basado en reglas de negocio
# Regla: Aprobado si:
# - Ingresos > 45K Y
# - Historial crediticio > 600 Y
# - Años de empleo > 2
y = np.where(
    (ingresos > 45) & 
    (historial_crediticio > 600) & 
    (anios_empleo > 2), 
    1, 0
)

# Dividir en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Normalizar datos
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Entrenar el perceptrón
model = Perceptron(learning_rate=0.01, n_iterations=1000)
model.fit(X_train_scaled, y_train)

# Evaluar el modelo
y_pred = model.predict(X_test_scaled)
accuracy = np.mean(y_test == y_pred)

print("=== SISTEMA DE APROBACIÓN DE PRÉSTAMOS CON PERCEPTRÓN ===")
print(f"Precisión del modelo: {accuracy:.2%}")

# Mostrar pesos y bias
print("\nPesos del modelo:")
print(f"w1 (Ingresos): {model.weights[0]:.4f}")
print(f"w2 (Historial crediticio): {model.weights[1]:.4f}")
print(f"w3 (Años de empleo): {model.weights[2]:.4f}")
print(f"Bias: {model.bias:.4f}")

# Mostrar qué es lo que más influye en la decisión
feature_names = ["Ingresos", "Historial crediticio", "Años de empleo"]
importance = abs(model.weights)
importance_normalized = importance / np.sum(importance)

print("\nImportancia relativa de cada factor:")
for i, feature in enumerate(feature_names):
    print(f"{feature}: {importance_normalized[i]:.2%}")

# Visualizar la evolución del aprendizaje
plt.figure(figsize=(10, 5))
plt.plot(model.accuracy_history)
plt.title("Evolución de la precisión durante el entrenamiento")
plt.xlabel("Iteraciones")
plt.ylabel("Precisión")
plt.grid(True)
plt.savefig('evolucion_aprendizaje.png')

# Ejemplo de uso con nuevos clientes
print("\n=== PRUEBA CON NUEVOS CLIENTES ===")

# Crear función para evaluar nuevos clientes
def evaluar_cliente(ingresos, historial_crediticio, anios_empleo):
    cliente = np.array([[ingresos, historial_crediticio, anios_empleo]])
    cliente_scaled = scaler.transform(cliente)
    
    # Calcular la suma ponderada
    u = model.weighted_sum(cliente_scaled)[0]
    decision = model.activate(np.array([u]))[0]
    
    print(f"Cliente: Ingresos=${ingresos}K, Historial={historial_crediticio}, Años empleado={anios_empleo}")
    print(f"Suma ponderada (u): {u:.4f}")
    print(f"Decisión: {'APROBADO' if decision == 1 else 'RECHAZADO'}")
    print("-" * 50)
    
    return decision

# Probar con algunos clientes
print("Cliente 1 (debe ser aprobado):")
evaluar_cliente(60, 720, 5)

print("Cliente 2 (debe ser rechazado):")
evaluar_cliente(30, 580, 1)

print("Cliente 3 (caso límite):")
evaluar_cliente(48, 610, 2.5)

# Visualizar la frontera de decisión (simplificada para 2 dimensiones)
plt.figure(figsize=(10, 8))

# Seleccionamos solo ingresos y historial crediticio para visualización
X_visual = X[:, :2]

# Crear un escalador específico para las 2 dimensiones que vamos a visualizar
visual_scaler = StandardScaler()
X_visual_scaled = visual_scaler.fit_transform(X_visual)

# Crear un modelo simplificado usando solo las dos primeras características
model_visual = Perceptron(learning_rate=0.01, n_iterations=1000)
X_train_visual = X_train[:, :2]
X_train_visual_scaled = visual_scaler.transform(X_train_visual)
model_visual.fit(X_train_visual_scaled, y_train)

# Crear una malla para visualizar la frontera de decisión
h = 0.02  # tamaño del paso en la malla
x_min, x_max = X_visual_scaled[:, 0].min() - 1, X_visual_scaled[:, 0].max() + 1
y_min, y_max = X_visual_scaled[:, 1].min() - 1, X_visual_scaled[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))

# Predecir para cada punto de la malla
Z = model_visual.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

plt.contourf(xx, yy, Z, alpha=0.3, cmap=plt.cm.Paired)
plt.scatter(X_visual_scaled[:, 0], X_visual_scaled[:, 1], c=y, cmap=plt.cm.Paired, edgecolors='k')

plt.xlabel('Ingresos (normalizado)')
plt.ylabel('Historial crediticio (normalizado)')
plt.title('Frontera de decisión del perceptrón para aprobación de préstamos')
plt.savefig('frontera_decision_prestamos.png')

print("\nImágenes guardadas: 'evolucion_aprendizaje.png' y 'frontera_decision_prestamos.png'")
print("\nCONCLUSIÓN:")
print("Este perceptrón implementa un sistema básico de decisión de préstamos")
print("basado en la fórmula u = x1*w1 + x2*w2 + x3*w3 + b, donde:")
print("- x1: Ingresos del solicitante")
print("- x2: Puntuación de historial crediticio")
print("- x3: Años de empleo")
print("- b: Sesgo (umbral base de aprobación)")
````

````python
import numpy as np
import matplotlib.pyplot as plt

class SimpleNeuron:
    def __init__(self):
        # Inicializar pesos y umbral en cero (explícitamente como flotantes)
        self.coeficientes = np.array([0.0, 0.0])  # c1, c2 como float
        self.umbral = 0.0                        # u como float
    
    def calcular(self, inputs):
        # Implementación de suma_ponderada = x1*c1 + x2*c2 + u
        suma_ponderada = np.dot(inputs, self.coeficientes) + self.umbral
        # Función de activación escalón
        return 1 if suma_ponderada > 0 else 0
    
    def aprender(self, datos_entrada, resultados_esperados, tasa_aprendizaje=0.1, iteraciones=100):
        # Entrenamiento simple del perceptrón
        for _ in range(iteraciones):
            for i in range(len(datos_entrada)):
                # Predicción
                resultado_obtenido = self.calcular(datos_entrada[i])
                # Actualización de pesos y umbral
                diferencia = resultados_esperados[i] - resultado_obtenido
                self.coeficientes += tasa_aprendizaje * diferencia * datos_entrada[i]
                self.umbral += tasa_aprendizaje * diferencia

# Datos para la compuerta lógica AND
entradas = np.array([
    [0, 0],  # caso 1
    [0, 1],  # caso 2
    [1, 0],  # caso 3
    [1, 1]   # caso 4
])

salidas = np.array([0, 0, 0, 1])  # resultados esperados para AND

# Crear y entrenar la neurona
neurona_simple = SimpleNeuron()
neurona_simple.aprender(entradas, salidas)

# Mostrar resultados
print("Coeficientes finales:", neurona_simple.coeficientes)
print("Umbral final:", neurona_simple.umbral)
print("\nResultados:")
print("0 AND 0 =", neurona_simple.calcular([0, 0]))
print("0 AND 1 =", neurona_simple.calcular([0, 1]))
print("1 AND 0 =", neurona_simple.calcular([1, 0]))
print("1 AND 1 =", neurona_simple.calcular([1, 1]))

# Visualización simple
plt.figure(figsize=(8, 6))
plt.scatter(entradas[:, 0], entradas[:, 1], c=salidas, s=100, cmap='plasma')

# Dibujar la frontera de decisión
limite_x_min, limite_x_max = -0.5, 1.5
limite_y_min, limite_y_max = -0.5, 1.5
malla_x, malla_y = np.meshgrid(np.arange(limite_x_min, limite_x_max, 0.01),
                     np.arange(limite_y_min, limite_y_max, 0.01))
valores_z = np.array([neurona_simple.calcular([x, y]) for x, y in zip(malla_x.ravel(), malla_y.ravel())])
valores_z = valores_z.reshape(malla_x.shape)
plt.contourf(malla_x, malla_y, valores_z, alpha=0.3)

plt.xlim(limite_x_min, limite_x_max)
plt.ylim(limite_y_min, limite_y_max)
plt.grid(True)
plt.title('Neurona simple implementando compuerta AND')
plt.xlabel('Entrada A')
plt.ylabel('Entrada B')
plt.show()
````

# Funciones de Pérdida en Perceptrones

## 📚 Introducción

Las funciones de pérdida (loss functions) son componentes fundamentales en el entrenamiento de modelos de machine learning, incluidos los perceptrones. Estas funciones cuantifican qué tan bien o mal está funcionando nuestro modelo, proporcionando una señal que guía el proceso de aprendizaje.

## 🎯 ¿Qué es una Función de Pérdida?

Una función de pérdida mide la discrepancia entre la predicción del modelo y el valor real esperado. En el contexto de perceptrones:

- **Objetivo**: Minimizar el error entre las predicciones y los valores reales
- **Uso**: Guía el algoritmo de optimización para ajustar los pesos y el sesgo
- **Resultado**: Un valor escalar que representa el "costo" o "error" del modelo

## 📊 Funciones de Pérdida para Perceptrones

### 1. Error Cuadrático (Mean Squared Error - MSE)

```python
def mse_loss(y_true, y_pred):
    return np.mean((y_true - y_pred) ** 2)
```

- **Características**: Penaliza errores grandes más severamente (cuadráticamente)
- **Matemáticamente**: $L = \frac{1}{n}\sum_{i=1}^{n}(y_i - \hat{y}_i)^2$
- **Uso**: Comúnmente utilizada en problemas de regresión

### 2. Error de Clasificación (Perceptron Loss)

```python
def perceptron_loss(y_true, y_pred_raw):
    # y_true: valores {-1, 1} para clasificación binaria
    # y_pred_raw: valores antes de la función de activación
    return np.maximum(0, -y_true * y_pred_raw)
```

- **Características**: Es cero cuando la predicción es correcta, positiva cuando es incorrecta
- **Matemáticamente**: $L = \max(0, -y \cdot \hat{y})$ donde $y \in \{-1, 1\}$
- **Uso**: La pérdida original del perceptrón, diseñada para clasificación binaria

### 3. Entropía Cruzada Binaria (Binary Cross-Entropy)

```python
def binary_cross_entropy(y_true, y_pred):
    epsilon = 1e-15  # Para evitar log(0)
    y_pred = np.clip(y_pred, epsilon, 1 - epsilon)
    return -np.mean(y_true * np.log(y_pred) + (1 - y_true) * np.log(1 - y_pred))
```

- **Características**: Penaliza fuertemente las predicciones confiadas pero incorrectas
- **Matemáticamente**: $L = -\frac{1}{n}\sum_{i=1}^{n}[y_i \log(\hat{y}_i) + (1-y_i) \log(1-\hat{y}_i)]$
- **Uso**: Popular en clasificación binaria cuando se usa una función de activación sigmoidea

## ⚙️ Entrenamiento con Funciones de Pérdida

### Proceso General

1. **Forward Pass**: Calcular la predicción del modelo
2. **Cálculo de Pérdida**: Medir el error usando la función de pérdida
3. **Backward Pass**: Calcular los gradientes (derivadas) de la pérdida respecto a los parámetros
4. **Actualización**: Ajustar los pesos para minimizar la pérdida

### Ejemplo: Entrenamiento de Perceptrón con MSE

```python
class PerceptronWithMSE:
    def __init__(self, n_inputs):
        self.weights = np.random.randn(n_inputs) * 0.01
        self.bias = 0
        
    def predict(self, X):
        return np.dot(X, self.weights) + self.bias
    
    def activate(self, z):
        return 1 if z > 0 else 0
    
    def compute_gradients(self, X, y, y_pred):
        # Gradiente de MSE: 2 * (y_pred - y) * x
        error = y_pred - y
        dw = 2 * np.dot(X.T, error) / len(y)
        db = 2 * np.mean(error)
        return dw, db
    
    def train(self, X, y, learning_rate=0.01, epochs=1000):
        losses = []
        
        for epoch in range(epochs):
            # Forward pass
            y_pred = self.predict(X)
            
            # Calcular pérdida
            loss = np.mean((y - y_pred) ** 2)  # MSE
            losses.append(loss)
            
            # Calcular gradientes
            dw, db = self.compute_gradients(X, y, y_pred)
            
            # Actualizar parámetros
            self.weights -= learning_rate * dw
            self.bias -= learning_rate * db
            
        return losses
```

## 📈 Evolución de la Pérdida Durante el Entrenamiento

Durante el entrenamiento, la función de pérdida típicamente sigue este patrón:

```
        │
Pérdida │    *
        │     *
        │      *
        │       *
        │        *
        │         *
        │          ***
        │             ******
        │                   **************
        └───────────────────────────────────►
                       Épocas
```

## 🔍 Comparación de Funciones de Pérdida

| Función | Ventajas | Desventajas | Mejor para |
|---------|----------|-------------|------------|
| **MSE** | Simple, derivable | Sensible a outliers | Regresión, valores continuos |
| **Perceptron Loss** | Directamente optimiza la clasificación | No derivable en 0 | Clasificación binaria básica |
| **Binary Cross-Entropy** | Mejor convergencia para clasificación | Requiere predicciones en [0,1] | Clasificación con probabilidades |

## 🛠️ Optimización de la Función de Pérdida

La elección de la función de pérdida depende de:

1. **Tipo de problema**: Clasificación vs. regresión
2. **Distribución de datos**: Balanceados vs. desbalanceados
3. **Sensibilidad a outliers**: Robustez requerida
4. **Interpretabilidad**: Facilidad para entender el significado del valor de pérdida

## 🔄 Relación entre Pérdida y Actualización de Pesos

En un perceptrón, los pesos se actualizan siguiendo:

$w_i = w_i - \alpha \frac{\partial L}{\partial w_i}$

Donde:
- $w_i$ es el peso i-ésimo
- $\alpha$ es la tasa de aprendizaje
- $\frac{\partial L}{\partial w_i}$ es el gradiente de la pérdida respecto al peso

## 📝 Conclusión

Las funciones de pérdida son el corazón del proceso de aprendizaje en perceptrones y redes neuronales. Proporcionan la señal de error que permite al modelo ajustarse y mejorar con el tiempo. La elección adecuada de la función de pérdida es crucial para el éxito del entrenamiento y el rendimiento del modelo resultante.


````python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from matplotlib import colors
import matplotlib.gridspec as gridspec

# Semilla para reproducibilidad
np.random.seed(42)

# Función para generar datos de dos clases
def generate_data(n_samples=100):
    # Generar primera clase (puntos en el cuadrante superior derecho)
    X_class1 = np.random.rand(n_samples//2, 2) * 0.5 + 0.5
    y_class1 = np.ones(n_samples//2)
    
    # Generar segunda clase (puntos en el cuadrante inferior izquierdo)
    X_class2 = np.random.rand(n_samples//2, 2) * 0.5
    y_class2 = np.zeros(n_samples//2)
    
    # Combinar ambas clases
    X = np.vstack([X_class1, X_class2])
    y = np.hstack([y_class1, y_class2])
    
    # Mezclar los datos
    indices = np.random.permutation(n_samples)
    X, y = X[indices], y[indices]
    
    return X, y

class Perceptron:
    def __init__(self, n_inputs):
        # Inicializar pesos y sesgo con valores pequeños aleatorios
        self.weights = np.random.randn(n_inputs) * 0.1
        self.bias = np.random.randn() * 0.1
        
        # Historial para la animación
        self.weight_history = [self.weights.copy()]
        self.bias_history = [self.bias]
        self.loss_history = []
        
    def forward(self, X):
        # Cálculo de la suma ponderada
        z = np.dot(X, self.weights) + self.bias
        return z
    
    def predict(self, X):
        # Predicción binaria usando la función escalón
        z = self.forward(X)
        return (z > 0).astype(int)
    
    def compute_loss(self, X, y):
        # Cálculo de pérdida usando MSE
        z = self.forward(X)
        y_pred = self.sigmoid(z)  # Usamos sigmoid para pérdida suave
        loss = np.mean((y - y_pred) ** 2)
        return loss
    
    def sigmoid(self, z):
        # Función de activación sigmoidea para valores suaves
        return 1 / (1 + np.exp(-z))
    
    def train_step(self, X, y, learning_rate=0.1):
        # Paso hacia adelante
        z = self.forward(X)
        y_pred = self.sigmoid(z)
        
        # Cálculo de errores y derivadas
        error = y_pred - y
        dw = np.dot(X.T, error) / len(y)
        db = np.mean(error)
        
        # Actualización de pesos y sesgo
        self.weights -= learning_rate * dw
        self.bias -= learning_rate * db
        
        # Guardar historial para animación
        self.weight_history.append(self.weights.copy())
        self.bias_history.append(self.bias)
        
        # Calcular y guardar la pérdida actual
        loss = self.compute_loss(X, y)
        self.loss_history.append(loss)
        
        return loss

# Generar datos
X, y = generate_data(200)

# Crear perceptrón
perceptron = Perceptron(n_inputs=2)

# Configuración del gráfico para la animación
fig = plt.figure(figsize=(15, 8))
gs = gridspec.GridSpec(2, 2, width_ratios=[2, 1])

# Panel izquierdo: Datos y frontera de decisión
ax1 = plt.subplot(gs[:, 0])
ax1.set_xlim(0, 1)
ax1.set_ylim(0, 1)
ax1.set_xlabel('Característica 1')
ax1.set_ylabel('Característica 2')
ax1.set_title('Datos y Frontera de Decisión')

# Mostrar clases con colores
scatter = ax1.scatter(X[:, 0], X[:, 1], c=y, cmap='coolwarm', edgecolors='k', alpha=0.7)
ax1.legend(*scatter.legend_elements(), title="Clases")

# Panel derecho superior: Gráfica de pérdida
ax2 = plt.subplot(gs[0, 1])
ax2.set_xlim(0, 80)
ax2.set_ylim(0, 0.3)
ax2.set_xlabel('Época')
ax2.set_ylabel('Pérdida')
ax2.set_title('Evolución de la Pérdida')
ax2.grid(True)
loss_line, = ax2.plot([], [], 'r-', lw=2)

# Panel derecho inferior: Pesos actuales
ax3 = plt.subplot(gs[1, 1])
ax3.set_xlim(-1, 1)
ax3.set_ylim(-1, 1)
ax3.set_xlabel('Peso 1')
ax3.set_ylabel('Peso 2')
ax3.set_title('Evolución de los Pesos')
ax3.set_aspect('equal')
ax3.grid(True)
weight_point = ax3.scatter([], [], color='blue', s=100, edgecolor='k')
weight_path, = ax3.plot([], [], 'b-', alpha=0.5)
ax3.add_patch(plt.Circle((0, 0), 1.0, fill=False, color='grey', linestyle='--'))

plt.tight_layout()

# Función para dibujar la frontera de decisión
def plot_decision_boundary(ax, weights, bias, color='black'):
    # Calculamos dos puntos de la línea de decisión
    if abs(weights[1]) < 1e-10:  # Evita división por cero
        x1 = -bias / weights[0]
        x2 = x1
        y1, y2 = 0, 1
    else:
        x1, x2 = 0, 1
        # Ecuación de la recta: w1*x + w2*y + b = 0 -> y = (-w1*x - b) / w2
        y1 = (-weights[0] * x1 - bias) / weights[1]
        y2 = (-weights[0] * x2 - bias) / weights[1]
    
    # Dibuja la línea de decisión
    line, = ax.plot([x1, x2], [y1, y2], color=color, lw=2)
    return line

# Inicializar la línea de decisión
decision_boundary_line = plot_decision_boundary(ax1, perceptron.weights, perceptron.bias)

# Función para la animación con manejo correcto de eliminación de líneas
def update(frame):
    global decision_boundary_line
    
    # Entrenar el perceptrón en cada frame
    if frame < 80:  # Limitamos a 80 épocas para la animación
        perceptron.train_step(X, y, learning_rate=0.1)
    
    # Actualizar la frontera de decisión
    weights = perceptron.weight_history[frame]
    bias = perceptron.bias_history[frame]
    
    # Eliminar la línea anterior de manera segura
    try:
        decision_boundary_line.remove()
    except:
        pass  # Si falla, ignoramos el error
    
    # Dibujar nueva frontera
    decision_boundary_line = plot_decision_boundary(ax1, weights, bias)
    ax1.set_title(f'Datos y Frontera de Decisión (Época {frame})')
    
    # Actualizar gráfica de pérdida si hay datos
    if frame > 0:
        loss_line.set_data(range(frame), perceptron.loss_history[:frame])
    
    # Actualizar gráfica de pesos
    weight_point.set_offsets([weights[0], weights[1]])
    
    # Actualizar trayectoria de pesos
    if frame > 0:
        w1_history = [w[0] for w in perceptron.weight_history[:frame+1]]
        w2_history = [w[1] for w in perceptron.weight_history[:frame+1]]
        weight_path.set_data(w1_history, w2_history)
    
    return [decision_boundary_line, loss_line, weight_point, weight_path]

# Crear animación (desactivar blit para evitar problemas)
# OJO QUE ESTÁ ADAPTADO PARA 200 FRAMES, BEJENLE SI NO LES FUNCAAAA
ani = FuncAnimation(fig, update, frames=range(99),
                   interval=100, blit=False, repeat=False)

# Pre-entrenar para poder guardar la animación de manera más confiable
print("Pre-entrenando el perceptrón...")
for _ in range(80):
    perceptron.train_step(X, y, learning_rate=0.1)

# Guardar animación usando método más confiable
print("Guardando animación...")
from matplotlib.animation import PillowWriter

# Personalizar la duración del GIF ajustando los parámetros
fps = 5  # Frames por segundo (valores más bajos = GIF más lento)
writer = PillowWriter(fps=fps)

# Guardar la animación como GIF
ani.save('perceptron_animation.gif', writer=writer)
print(f"Animación guardada como 'perceptron_animation.gif' con {fps} FPS")

# Mostrar la figura
plt.close()  # Cerrar la figura original para evitar duplicados
plt.show()

# Mostrar el GIF en el notebook (si estás en un entorno que lo soporte)
try:
    from IPython.display import Image, display
    # Mostrar directamente en el notebook
    display(Image('perceptron_animation.gif'))
    print("GIF mostrado exitosamente en el notebook")
except ImportError:
    print("Para visualizar el GIF en el notebook, instala IPython con: pip install ipython")
except Exception as e:
    print(f"Error al mostrar el GIF: {e}")
    print("El GIF se ha guardado pero no se puede mostrar directamente en este entorno")
````
